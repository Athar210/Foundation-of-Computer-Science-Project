```python
import pandas as pd

CA = pd.read_csv("trendingYT/CAvideos.csv.zst", encoding="latin1")
CA["country"] = "CA"

DE = pd.read_csv("trendingYT/DEvideos.csv.zst", encoding="latin1")
DE["country"] = "DE"

FR = pd.read_csv("trendingYT/FRvideos.csv.zst", encoding="latin1")
FR["country"] = "FR"

GB = pd.read_csv("trendingYT/GBvideos.csv.zst", encoding="latin1")
GB["country"] = "GB"

IN = pd.read_csv("trendingYT/INvideos.csv.zst", encoding="latin1")
IN["country"] = "IN"

JP = pd.read_csv("trendingYT/JPvideos.csv.zst", encoding="latin1")
JP["country"] = "JP"

KR = pd.read_csv("trendingYT/KRvideos.csv.zst", encoding="latin1")
KR["country"] = "KR"

MX = pd.read_csv("trendingYT/MXvideos.csv.zst", encoding="latin1")
MX["country"] = "MX"

RU = pd.read_csv("trendingYT/RUvideos.csv.zst", encoding="latin1")
RU["country"] = "RU"

US = pd.read_csv("trendingYT/USvideos.csv.zst", encoding="latin1")
US["country"] = "US"
```

```python
# question 5
df_all["like_ratio"] = df_all["likes"] / df_all["dislikes"].replace(0, pd.NA)
df_all[["likes", "dislikes", "like_ratio"]].head()
```

```python
# question 6
df_all["publish_time"] = pd.to_datetime(df_all["publish_time"])
df_all["publish_interval"] = df_all["publish_time"].dt.floor("10min")
df_all[["publish_time", "publish_interval"]].head()
```

```python
# question 7
interval_stats = (
    df_all
    .groupby("publish_interval")
    .agg(
        video_count=("video_id", "count"),
        avg_likes=("likes", "mean"),
        avg_dislikes=("dislikes", "mean")
    )
    .reset_index()
)
interval_stats.sort_values("video_count", ascending=False).head(10)
```

```python
# question 8
tag_counts = (
    df_all.assign(tag_list=df_all["tags"].str.split("|"))  # split tags
          .explode("tag_list")                             # one tag per row
          .assign(tag_list=lambda x: x["tag_list"].str.strip().str.strip('"').str.strip("'"))  # clean
          .query("tag_list != '' and tag_list.str.lower() != '[none]'")  # remove empty/[none]
          .groupby("tag_list", as_index=False)["video_id"].count()       # count videos
          .rename(columns={"video_id": "video_count"})
          .sort_values("video_count", ascending=False)
)

tag_counts.head(10)
```

```python
# question 9
top_tags = tag_counts.head(10)
top_tags
```

```python
# question 10
# Step 1: Explode and clean tags (once)
df_tags_exploded = (
    df_all.assign(tag_list=df_all["tags"].str.split("|"))  # split tags
          .explode("tag_list")                             # one tag per row
          .assign(tag_list=lambda x: x["tag_list"].str.strip().str.strip('"').str.strip("'"))  # clean
          .query("tag_list != '' and tag_list.str.lower() != '[none]'")  # remove empty/[none]
)

# Step 2: Compute average like_ratio per (tag, country)
tag_country_ratio = (
    df_tags_exploded.groupby(["tag_list", "country"], as_index=False)["like_ratio"]
                     .mean()
                     .rename(columns={"like_ratio": "avg_like_ratio"})
)

tag_country_ratio.head(10)
```

```python
# question 11
# Ensure trending_date is datetime
df_all["trending_date"] = pd.to_datetime(df_all["trending_date"], format="%y.%d.%m")  # YouTube format: YY.DD.MM

# Find video with max views per (trending_date, country)
top_videos_daily = (
    df_all.loc[df_all.groupby(["trending_date", "country"])["views"].idxmax()]
    .reset_index(drop=True)
)

top_videos_daily.head()
```

```python
# question 14
import json
import glob
import os
import pandas as pd

# Read all category JSON files into a single DataFrame
category_files = glob.glob(r"trendingYT\*_category_id.json")  # all files
category_list = []

for file in category_files:
    country = os.path.basename(file).split("_")[0].upper()  # extract country code
    with open(file, encoding="utf-8") as f:
        data = json.load(f)
    
    df = pd.DataFrame([{
        "category_id": int(item["id"]),
        "category_title": item["snippet"]["title"],
        "country": country
    } for item in data.get("items", [])])
    
    category_list.append(df)

# Combine all countries into one DataFrame
categories_df = pd.concat(category_list, ignore_index=True)

# Inspect
categories_df.head()
```

```python
# Merge and fill missing categories
df_all_cat = df_all.merge(categories_df, on=["country", "category_id"], how="left")
df_all_cat["category_title"] = df_all_cat["category_title"].fillna("Not assignable")

# Count Not assignable videos per country
not_assignable_counts = (
    df_all_cat[df_all_cat["category_title"] == "Not assignable"]
    .groupby("country")["video_id"]
    .count()
    .to_dict()
)

not_assignable_counts
```
